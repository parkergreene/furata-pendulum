# Tuning PID Controller

https://github.com/ShawnHymel/pendulum-pid/blob/master/pendulum-pid.ipynb

```python
kP = 1 
kI = 0 
kD = 0 
bias = 0

setpoint = 90
position = 89
time_interval = 2

error_prev = 0
integral = 0
move_deg = 0


error  = setpoint - position
integral = integral + (error * time_interval)
derivative = (error - error_prev) / time_interval
move_deg = (kP * error) + (kI * integral) + (kD * derivative) + bias

error_prev = error

print(move_deg)
```
